```python
#import 
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Set funcitons and their quations to be used later
def charge_on_C(qa,qb):
    qc = -(qa+qb)/3
    return qc    

#Set the function shell charge 
def shell_charge(qa,qb):
    qsc = -(qa+qb+3*0.378)/3
    return qsc

#Set the function of the spring constant
def spring(qsc):
    ksc = qsc**2/0.078
    return ksc
```

```python
#Set energy values of atom B to go between parameters 1.5->2.5
qa = np.arange(0.5,1.6,0.1)
qa
```

```python
#Set energy values of atom B to go between parameters 1.5->2.5
qb = np.arange(1.5,2.6,0.1)
qb
```

```python
#Finding the charge on C(the halide)
for A in qa:
    for B in qb:
        print (A,B)
        print charge_on_C(A,B)
        
```

```python
for A in qa:
    for B in qb:
        print (A,B)
        print shell_charge(A,B)
```

```python
#Finding the Shell-charge on C
#Creating a loop for values in aq and in qb
for A in qa:
    for B in qb:
        qsc = shell_charge(A,B)
        
        print (A,B)
        print spring(qsc)
```

```python
#turn the above data into a list of [qa,qb,qsc,ksc]
mylist = []

for A in qa:
    for B in qb:
        qsc = shell_charge(A,B)
        ksc = spring(qsc)
        
        mylist.append([A,B,qsc,ksc])

mylist = [item[0:4] for item in mylist]
        
```

```python
#Check that the above has worked
for item in mylist:
    print item
```
