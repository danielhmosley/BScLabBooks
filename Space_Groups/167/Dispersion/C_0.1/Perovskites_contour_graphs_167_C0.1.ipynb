```python
import numpy as np
import matplotlib.pyplot as plt

# setup some generic data
N = 37
x, y = np.mgrid[:N, :N]
Z = (np.cos(x*0.7) + np.sin(y*0.3))

fig, ax = plt.subplots(1)

# plot just the positive data and save the
# color "mappable" object returned by ax1.imshow
pos = ax.imshow(Z, cmap='Blues', interpolation='bilinear')

# add the colorbar using the figure's method,
# telling which mappable we're talking about and
# which axes object it should be near
fig.colorbar(pos, ax=ax)

plt.show()
```

```python
list1 = []

file = open("spacegroup167_C0_1.txt","r")


for line in file.readlines():
    list1.append(line.split()[0])
file.close()

two_d = np.zeros(shape=(11,11))

for i in range(11):
    for j in range(11):
        two_d[i,j] = list1[i*11 + j]
    
two_d

```

```python
fig, ax = plt.subplots(1)

# plot just the positive data and save the
# color "mappable" object returned by ax1.imshow
# use vmin and vmax to select the range of the colour bar
pos = ax.imshow(two_d, cmap='cool', interpolation='bilinear', vmin=0, vmax=11)

# add the colorbar using the figure's method,
# telling which mappable we're talking about and
# which axes object it should be near
fig.colorbar(pos, ax=ax)
ax.set_xlabel('$Q_B (-e)$',size=22)
ax.set_ylabel('$Q_A (-e)$',size=22)


labelsB = ['0.4', '0.5', '0.7', '0.9', '1.1', '1.3', '1.5']
labelsA = ['1.4', '1.5', '1.7', '1.9', '2.1', '2.3', '2.5']

ax.set_xticklabels(labelsA)
ax.set_yticklabels(labelsB)

plt.tight_layout()
plt.savefig('Spacegroup_62_energies.png')
plt.show()
```
